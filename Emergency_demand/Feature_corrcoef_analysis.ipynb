```python
# import line
import pandas as pd
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# 데이터 로드
data = pd.read_csv('train_data_diff.csv', low_memory=False)
len(data.columns)
```

```python
# 결측치 제거
data_cleaned = data.dropna()
data_cleaned = data_cleaned.drop('Unnamed: 0', axis=1)
```

```python
# 메타 데이터 제외
data_cleaned.drop('DONG', axis=1, inplace=True)
data_cleaned.drop('DATETIME', axis=1, inplace=True)

# 값이 0만 존재하므로 제외
print(np.unique(data_cleaned['PAT_SEXDSTN_CD_0590040']))
data_cleaned.drop('PAT_SEXDSTN_CD_0590040', axis=1, inplace=True)
```

```python
# 데이터 표준화
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_cleaned)

# 열별 상관계수 계산
corr_matrix = np.corrcoef(data_scaled, rowvar=False) 
corr_matrix = np.nan_to_num(corr_matrix) # nan->0
```

```python
cols = data_cleaned.columns

for i, r in enumerate(corr_matrix):
    df_r = pd.DataFrame({'r':r, '|r|':abs(r)}, index=cols)
    
    # r이 1인 값 제외
    df_r = df_r[df_r.r!=1.]
    
    # |r| 기준 내림차순 정렬
    df_r.sort_values(by='|r|', ascending=False, inplace=True)
    
    # col.csv 형태로 저장
    #df_r.to_csv(f'corrcoef/{cols[i]}.csv')
    
    # df_r 출력
    print(cols[i])
    display(df_r)
    print('\n\n')
```
